# Conclusions Drawn From Analysis

1. Data imbalance is quiet high in TARGET variable. And the imbalance ratio is 10.55

2. The number of female defaulters is slightly higher than male defaulters, and the same trend is observed for non-defaulters as well where females outnumber males.

3. In both cases, i.e., for defaulters and non-defaulters, the number of revolving loans is significantly lower compared to cash loans.

4. Working people have the highest number of default cases compared to other professions, and the same trend is observed for non-defaulters, where working people have the highest number of non-default cases.

5. Customers with secondary or secondary special education level have the highest number of default cases compared to other education levels, and the same trend is observed for non-defaulters, where customers with secondary or secondary special education level have the highest number of non-default cases.

6. Young people are more likely to default than other age groups, whereas senior citizens are less likely to default than other age groups, for both defaulters and non-defaulters. However, there is not much difference in the likelihood for non-defaulters in the age groups.

7. We can observe a concentration of values on lower income and lower credit of loans for defaulters, indicating that higher income is associated with higher loan amounts. This trend is observed for both genders. However, for non-defaulters, no clear pattern can be identified.
